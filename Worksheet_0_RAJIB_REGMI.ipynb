```python
print("TASK FIRST OF RECURSION")
def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list): 
            total += sum_nested_list(element)
        else:
            total += element  # Add the number to the total
    return total
nested_list1 = [1, [2, [3, 4], 5], 6, [7, 8]]
nested_list2 = [1, [2, 3], [4, [5]]]
print("sum_nested_list1:",sum_nested_list(nested_list1))  
print("sum_nested_list2:",sum_nested_list(nested_list2))
```

```python
def generate_permutations(s):
    if len(s) <= 1:
        return [s]
    permutations = set()  # Use set to avoid duplicate permutations
    for i, char in enumerate(s):
        remaining = s[:i] + s[i+1:]
        for perm in generate_permutations(remaining):
            permutations.add(char + perm)
    
    return list(permutations)

# Test the function
print(generate_permutations("abc"))  
print(generate_permutations("aab"))  
```

```python
print("TASK THIRD OF RECURSION")

def calculate_total_size(directory):
    total_size = 0

    for key, value in directory.items():
        if isinstance(value, dict):  # If the value is a subdirectory
            total_size += calculate_total_size(value)  # Recursively calculate the size of the subdirectory
        else:  # If the value is a file, add its size
            total_size += value

    return total_size

# Example directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    }, "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        }, "file6.txt": 150
    }
}

# Call the function and print the result
print("Total directory size:", calculate_total_size(directory_structure))
```

```python
print("TASK FIRST OF DYNAMIC PROGRAMMING")
def min_coins(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0
    
    # Iterate over all amounts from 1 to the target amount
    for i in range(1, amount + 1):
        # For each coin, check if it can be used to form the amount i
        for coin in coins:
            if i >= coin:
                dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

# Test with coins = [1, 2, 5] and amount = 11
coins = [1, 2, 5]
amount = 11
result = min_coins(coins, amount)

print("The minimum number of coins needed:", result)  

```

```python
print("TASK SECOND OF DYNAMIC PROGRAMMING")

def longestcommon_subsequence(s1, s2):
    dp = [[0] * (len(s2) + 1) for _ in range(len(s1) + 1)]
    
    for i in range(1, len(s1) + 1):
        for j in range(1, len(s2) + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    
    return dp[len(s1)][len(s2)]

s1 = "abcde"
s2 = "ace"
print("The Longestcommon_subsequence is:", longestcommon_subsequence(s1, s2))
```

```python
print("TASK THIRD OF DYNAMIC PROGRAMMING")

def knapsack(weights, values, capacity):
    n = len(weights)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]  # Initialize a DP table
    
    # Fill the DP table using the bottom-up approach
    for i in range(1, n + 1):  # For each item
        for w in range(1, capacity + 1):  # For each weight capacity
            if weights[i - 1] <= w:  # If the item can fit in the knapsack
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]  # Item cannot be included, so we keep the previous value
    
    return dp[n][capacity]

# Example input
weights = [1, 3, 4, 5]  
values = [1, 4, 5, 7]  
capacity = 7            

# Call the function
result = knapsack(weights, values, capacity)

# Output the result
print("The Maximum value is:", result)  
```
